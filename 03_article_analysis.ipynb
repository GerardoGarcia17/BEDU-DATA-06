### Analisis de datos

Sabes cuales de los diferentes sitios web (sources) tienen mas interacciones

```python
#Seleccionamos las columnas base para nuestro analisis
COLUMNS = [
    'source_name',
    'author',
    'engagement_reaction_count',
    'engagement_comment_count',
    'engagement_share_count',
    'engagement_comment_plugin_count'
]

#Limpiar todas aquellas filas que tienen un valor nulo en las columnas seleccionadas
data = data[COLUMNS].dropna(axis=0, how='any')
```

```python
#Agrupamos nuestro dataframe por la columna source_name y sumamos cada fila por su valor en la columna engagement_reaction_count
grouped_sum = data.groupby('source_name')['engagement_reaction_count'].sum()
grouped_sum.to_frame()
```

```python
grouped_sum = data.groupby('source_name')['engagement_share_count'].sum()
grouped_sum.to_frame()
```
